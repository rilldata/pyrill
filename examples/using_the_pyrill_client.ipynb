# PyRill Client - Jupyter Notebook Example

This notebook demonstrates using the PyRill client in a Jupyter environment.

## Important Note on Architecture

The RillClient is a **thin client** focused on data fetching only:
- ✅ Fetches data from Rill Data API (organizations, projects, resources)
- ✅ Handles authentication
- ✅ Optional caching

```python
from pyrill import RillClient
```

## Initialize Client

The client will use the `RILL_USER_TOKEN` from your environment.

```python
client = RillClient(org='demo', project='rill-openrtb-prog-ads')
print("Client initialized successfully!")
```

## Check Authentication

```python
user = client.auth.whoami()
user
```

## List Organizations

```python
orgs = client.organizations.list()
orgs
```

## List Projects

Let's get projects for the first organization:

```python
org_name = orgs[0].name
projects = client.projects.list(org_name=org_name)
projects
```

## Get Runtime Resources

Query runtime resources for a specific project:

```python
org_name = orgs[0].name
project_name = projects[0].name
resources = client.projects.get_resources(org_name, project_name)
resources
```
